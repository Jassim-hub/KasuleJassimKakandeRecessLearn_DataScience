```python
# DATAFRAMES
# (For this section, you will use the ‘AfricaCupOfNations’ dataset provided to you)
# 1.	Write a Pandas program to read the given csv file.
import pandas as pd
import numpy as np
df = pd.read_csv("AfricaCupofNationsMatches.csv")
print("\nDataFrame:")
print(df)
# 2.	Write a Pandas program to get the first 7 rows of your data frame.
print("\nFirst 7 rows of the DataFrame:")
print(df.head(7))
# 3.	Write a Pandas program to select the ‘HomeTeam’, ‘AwayTeam’, ‘HomeTeamGoals’ and ‘AwayTeamGoals’ columns from your data frame.
print("\nSelected columns:")
print(df[["HomeTeam", "AwayTeam", "HomeTeamGoals", "AwayTeamGoals"]])
# 4.	Write a Pandas program to select the rows where Egypt appears.
print("\nRows where Egypt appears:")
print(df[df["HomeTeam"] == "Egypt"])
```

```python
#5.	Write a Pandas program to count the number of rows and columns of your data frame.
print("\nNumber of rows and columns:")
print(df.shape)
#6.	Write a Pandas program to select the rows where the ‘Attendance’ is missing.
print("\nRows where Attendance is missing:")
print(df[df["Attendance"].isnull()])
```

```python
#7.	Write a Pandas program to select the rows where the ‘HomeTeamGoals’ are between 3 and 6 inclusive.
print("\nRows where HomeTeamGoals are between 3 and 6 inclusive:")
print(df[(df["HomeTeamGoals"] >= 3) & (df["HomeTeamGoals"] <= 6)])
#8.	Write a Pandas program to change the “AwayTeamGoals” in the 3rd row to 10.
print("\nChanging AwayTeamGoals in the 3rd row to 10:")
df.at[2, "AwayTeamGoals"] = 10
#9.	Write a Pandas program to sort the DataFrame first by ‘HomeTeam’ in ascending order, then by ‘HomeTeamGoals’ in descending order.
print("\nDataFrame sorted by HomeTeam ascending and HomeTeamGoals descending:")
sorted_df = df.sort_values(by=["HomeTeam", "HomeTeamGoals"], ascending=[True, False])
print(sorted_df)
```

```python
#10.Write a Pandas program to get list from DataFrame column headers.
print("\nList of DataFrame column headers:")
print(df.columns.tolist())
#11.Write a Pandas program to append a column of your choice to your DataFrame
print("\nAppending a new column 'MatchYear' to the DataFrame:")
df['MatchYear'] = 2025
print(df.head())
print(df.shape)
```

```python
#12. Write a Pandas program to add 2 rows to your DataFrame.
print("\nAdding 2 new rows to the DataFrame:")
# Create two new rows as dictionaries (ensure all columns are present)
new_rows = [
    {
        "HomeTeam": "TeamA",
        "AwayTeam": "TeamB",
        "HomeTeamGoals": 2,
        "AwayTeamGoals": 1,
        "Attendance": 15000,
        "MatchYear": 2025
    },
    {
        "HomeTeam": "TeamC",
        "AwayTeam": "TeamD",
        "HomeTeamGoals": 3,
        "AwayTeamGoals": 2,
        "Attendance": 18000,
        "MatchYear": 2025
    }
]
# Convert to DataFrame and append
new_df = pd.DataFrame(new_rows)
df = pd.concat([df, new_df], ignore_index=True)
print(df.tail(5))
print(df.shape)
```

```python
#13.Write a Pandas program to change the country ‘Uganda’ to ‘China’ in the ‘AwayTeam’ column of the DataFrame.
# print(df['AwayTeam'].unique())
print(df[df['AwayTeam'].str.strip().str.lower() == 'uganda'])
```

```python
# Remove leading and trailing whitespaces from all string columns in the DataFrame
str_cols = df.select_dtypes(include=['object']).columns
for col in str_cols:
    df[col] = df[col].str.strip()
#13.Write a Pandas program to change the country ‘Uganda’ to ‘China’ in the ‘AwayTeam’ column of the DataFrame.    
df['AwayTeam'] = df['AwayTeam'].replace('Uganda', 'China')
print(df[df['AwayTeam'] == 'China'])


```

```python
#14. Write a Pandas program to reset index in your DataFrame.
df = df.reset_index(drop=True)
print("\nIndex has been reset:")
print(df.head())
```

```python
#15.Write a Pandas program to check whether the ‘Stadium’ column is present in your DataFrame or not.
if 'Stadium' in df.columns:
    print("The 'Stadium' column is present in the DataFrame.")
else:
    print("The 'Stadium' column is not present in the DataFrame.")
```

```python
# 16.Write a Pandas program to convert the datatype of the ‘AwayTeamGoals’. (int to float)
df['AwayTeamGoals'] = df['AwayTeamGoals'].astype(float)
print(df.dtypes)
```

```python
#17.Write a Pandas program to remove the last 10 rows from your DataFrame.
df = df.iloc[:-10]
print("\nDataFrame after removing the last 10 rows:")
print(df.tail())
print("New shape:", df.shape)
```

```python
#18.Write a Pandas program to iterate over rows in your DataFrame.
for index, row in df.iterrows():
    print(f"Row {index}: HomeTeam = {row['HomeTeam']}, AwayTeam = {row['AwayTeam']}")
```

```python
# 19. Write a Pandas program to change the order of your DataFrame columns.
# Example: Move 'AwayTeam' to be the first column, followed by the rest
cols = list(df.columns)
if 'AwayTeam' in cols:
    cols.insert(0, cols.pop(cols.index('AwayTeam')))
df = df[cols]
print(df.head())
```

```python
#20.Write a Pandas program to delete DataFrame row(s) whose value is 0 in the ‘HomeTeamGoals’ columns.
df = df[df['HomeTeamGoals'] != 0]
print(df['HomeTeamGoals'].to_string(index=False))
```
